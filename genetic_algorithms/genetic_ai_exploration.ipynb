```python
%load_ext autoreload
%autoreload 2
from pacman import runGames, readCommand, change_display
import numpy as np
from evolve import genetic_algorithm_evolution
import copy
from tqdm import tqdm
import pickle
from IPython.display import clear_output


use_saved_agents = False # Choose if you want to use the saved agents or load random ones
agents_path = 'agents_top_-440.4.pkl' # Choose the agents to load

agents_to_generate = 100 # Choose the number of agents to generate

arguments = ['-l', 'evolutionTerain', '-p', 'NeuralAgent', '-q', '-e', 'save_best', '--pc', str(agents_to_generate), '-k', '2']

args = readCommand( arguments)

if args['evolution'] == 'save_best':

    if use_saved_agents:
        with open('AgentsFolder\\' + agents_path, 'rb') as f:
            agents = pickle.load(f)
    else:
        agents = copy.deepcopy(args['pacman'])

    zoom = args['zoom']
    del args['zoom']
```

```python
from concurrent.futures import ProcessPoolExecutor
from functools import partial


epochs = 100
random_factor = 0.4
percentage_reproduce = 0.05

max_score = -1000
agents_path = 'agents_top_it_6' # Choose the name of the file to save the best agents


for epoch in range(epochs):

    print('\nEpoch: ', epoch)
    
    scores = []
    
    for agent in tqdm(agents):
            args['pacman'] = agent
            scores.append(runGames( **args ))

    scores = np.array(scores)
    average = np.mean(scores)
    
    print('\nAverage Score:', average, '\n')

    if average > max_score:
        max_score = average
        print('New best score: ', max_score)
        agents_path_full = agents_path + '.pkl'
        with open('AgentsFolder\\' + agents_path_full, 'wb') as f:
            pickle.dump(agents, f)

    if epoch % 1 == 0:
        max_score_index = np.argmax(scores)
        args['display'] = change_display(False, zoom)
        args['pacman'] = agents[max_score_index]
        runGames(**args)
        args['display'] = change_display(True, zoom)
        
    if epoch % 5 == 0:
        clear_output(wait=True)


    agents = genetic_algorithm_evolution(agents, scores, agents_to_generate, random_factor=random_factor, percentage=percentage_reproduce)
```
